```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Import csv and plot them on the same graph
df_p_01 = pd.read_csv('singles_data_stab_fez_whole_pi_over_2_p_0.1.csv')
df_p_001 = pd.read_csv('singles_data_stab_fez_whole_pi_over_2_p_0.05.csv')
df_p_0001 = pd.read_csv('singles_data_stab_fez_whole_pi_over_2_p_0.01.csv')
df_p_00001 = pd.read_csv('singles_data_stab_fez_whole_pi_over_2_p_0.001.csv')

plt.figure(figsize=(8, 5))
plt.yscale('log')

plt.errorbar(
    df_p_01['lengths'],
    df_p_01['mmi_singles'],
    yerr=df_p_01['mmi_singles_err'],
    label=r'$p_2 = 0.1$'
)

plt.errorbar(
    df_p_001['lengths'],
    df_p_001['mmi_singles'],
    yerr=df_p_001['mmi_singles_err'],
    label=r'$p_2 = 0.05$'
)

plt.errorbar(
     df_p_0001['lengths'],
     df_p_0001['mmi_singles'],
     yerr=df_p_0001['mmi_singles_err'],
     label=r'$p_2 = 0.01$'
)

plt.errorbar(
    df_p_00001['lengths'],
    df_p_00001['mmi_singles'],
    yerr=df_p_00001['mmi_singles_err'],
    label=r'$p_2 = 0.001$'
)

plt.legend(loc='lower right')
plt.xlabel('Circuit Length')
plt.ylabel('Mean Mutual Information')

plt.tight_layout()

# X-axis every 5 units (5, 10, 15, ...)
plt.xticks(5 * np.arange(0, 21))

# Export image as png
plt.savefig('peak_shift_comparison.png', dpi=900)
plt.show()
```
